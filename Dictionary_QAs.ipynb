```
# Question 1: Write a function that takes a dictionary and a key as input and returns the corresponding value.
Sample Input: my_dict = {'a': 1, 'b': 2, 'c': 3}, key = 'b'
Sample Output: 2
```



````python
# ans:
def dict_value(self,x):
  if x in self:
    return self[x]
  else:
    return None
print(dict_value({'a': 1, 'b': 2, 'c': 3},'b'))
````

```
# Question 2 : Write a function that updates a dictionary with a new key-value pair.
Sample Input: my_dict = {'a': 1, 'b': 2}, key = 'c', value = 3
Sample Output:{'a': 1, 'b': 2, 'c': 3}
```



````python
#ans:
def updates(self,key,value):
  self[key] = value
  return self
print(updates({'a': 1, 'b': 2}, key = 'c', value = 3))
````

```
# Question 3 : Write a function that checks if a key exists in a dictionary.
Sample Input: my_dict = {'a': 1, 'b': 2, 'c': 3}, key = 'd'
Sample Output: False
```



````python
# ans :
def checking_key_exists(self,key):
  if key in self:
    return key
  else:
    return f'key {key} does not exists'
print(checking_key_exists({'a': 1, 'b': 2, 'c': 3},'d'))
````

```
# Question 4 : Write a function that calculates the sum of all values in a dictionary.
Sample Input: my_dict = {'a': 1, 'b': 2, 'c': 3}
Sample Output: 6
```



````python
#ans:
def calculate_dict_sum(self):
  dict_values = 0
  for i in self.values():
     dict_values += i
  return dict_values
print(calculate_dict_sum({'a': 1, 'b': 2, 'c': 3}))
````

```
# Question 5 : Write a function that returns a list of all keys in a dictionary.
Sample Input: my_dict = {'a': 1, 'b': 2, 'c': 3}
Sample Output: ['a', 'b', 'c']
```



````python
# 5th ans:
def list_dict_keys(self):
  return list(self.keys())
print(list_dict_keys({'a': 1, 'b': 2, 'c': 3}))
````

````python
# 5th ans:
def list_keys(self):
  new_list = ''
  for i in self.keys():
    new_list += i
  return list(new_list)
print(list_keys({'a': 1, 'b': 2, 'c': 3}))
````

```
# Question 7 : Write a function that returns the number of key-value pairs in a dictionary.
Sample Input: my_dict = {'a': 1, 'b': 2, 'c': 3}
Sample Output: 3
```



````python
# ans:
def dict_key_value_pairs(self):
  return len(self)
print(dict_key_value_pairs({'a':1,'b':2,'c':3}))
````

```
# Question 8 : Write a function that merges two dictionaries into one.
Sample Input: dict1 = {'a': 1, 'b': 2}, dict2 = {'c': 3, 'd': 4}
Sample Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
```



````python
def merging_dict(x,y):
  merged_dict = x.copy()
  merged_dict.update(y)
  return f'merged : {merged_dict}'
print(merging_dict({'a': 1, 'b': 2},{'c': 3, 'd': 4}))
````

```
# Question 9 : Write a function that sorts a dictionary by its values in ascending order.
Sample Input: my_dict = {'a': 5, 'b': 2, 'c': 7}
Sample Output: {'b': 2, 'a': 5, 'c': 7}
```



````python
# ans:
def dictionary_sort(self):
  return dict(sorted(self.items(),key = lambda item: item[1]))
print(dictionary_sort({'a':5,'b':2,'c':7}))
````

```
# Question 10 : Write a function that returns the key in a dictionary with the maximum value.
Sample Input: my_dict = {'a': 5, 'b': 2, 'c': 7}
Sample Output: 'c'
```



````python
# ans:
def max_value_dict(self):
  max_num = max(self,key = self.get)
  return f'max_value is {max_num}'
print(max_value_dict({'a': 5, 'b': 2, 'c': 7}))
````
