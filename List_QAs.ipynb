```
# Question 1: Write a Python program to count the number of elements in a list without Using len.
sample input :
var = [10,11,12,13]
sample output :
result : 4
```



````python
# ans:
def len_fun(self):
  list_count = 0
  for i in self:
    if i:
      list_count += 1
  return f'total number of elements in list is :{list_count}'
print(len_fun([10,11,12,13]))
````


# Question 2: Write a Python program to find the maximum and minimum values in a given list.
sample input :
var = [10,11,12,13,14,15]
sample output :
result : max = 15 , min = 10







````python
def max_min(self):
  max = self[0]
  min = self[0]
  for i in self:
    if i<min:
      min = i
    elif i > max:
      max = i
    else:
      pass
  return max , min
print(max_min([10,11,12,13,14,15]))
````

```
# Question 4: Write a Python program to find the average of all the elements in a list.
sample input :
var = [1,2,3,4,2,5]
sample output :
result : 2.833
```



````python
def average_elements(self):
  sum_ele = sum(self)
  total_ele = len(self)
  average = sum_ele / total_ele
  return average
print(average_elements([1,2,3,4,2,5]))
````

```
# Question 7: Write a Python program to find which number is palindrome in below given list .
sample input :
var = [134,121,155]
sample output :
result : 134 is not a palindome , 121 is a palindrome
```



````python
def num_palindrome(self):
  var = str(self)
  return var == var[::-1]
va = [134,121,155]
for i in va:
  if num_palindrome(i):
    print(f'{i} is a palindrome')
  else:
    print(f'{i} is not a palindrome')

````

```
# Question 8: Write a Python program to find the common elements between two lists.
sample input :
var1 = [1,2,3,4,10]
var2 = [10,11,12,14]
sample output :
result : 10
```



````python
# ans :
def common_elements (x,y):
  set1 = set(x)
  set2 = set(y)
  common_element = set1.intersection(set2)
  return  f'result:{list(common_element)}'
print(common_elements( [1,2,3,4,10],[10,11,12,14]))
````

```
# Question 9: Write a Python program to remove the duplicates from a list
sample input :
var = [10,11,12,10,12,24]
sample output :
result : [10,11,12,24]
```



````python
# ans:
def remove_duplicates(self):
  dict1 = []
  for i in self:
   if i not in dict1:
     dict1 += [i]
  return dict1

print(remove_duplicates([10,11,12,10,12,24]))
````
