````python
# Question 1: Write a Python program to count the number of characters in a string without using len function.
sample input :
var = 'python java'
sample output :
result : 11




def len(self):
    count = 0
    for i in self:
        if i :
            count += 1
    return f'len of var is :{count} characters'
print(len('python java'))
````

````python
# Question 2: Write a Python program to reverse a string without using [::-1]
sample input :
var = 'python'
sample output :
result : nohtyp

def rev(self):
    new_str = ''
    for i in self:
        new_str = i + new_str
    return new_str
print(rev('python'))
````

````python

# Question 3: Write a Python program to check if a string is a palindrome don't use [::-1].
sample input :

var = 'NUN'
sample output :

result : It is a Palindrome
palindrome concept = https://www.dictionary.com/e/palindromic-word/#:~:text=A%20palindrome%20is%20a%20word,roots%20in%20the%20early%201600s.




def palindrome(self):
    New_str = ''
    for i in self:
        New_str = i + New_str
    if New_str == self:
        return " It is a palindrome"
    else:
        return " NO, It is not a palindrome"
print(palindrome("NUN"))
````

````python


# [Question 4: Write a Python program to find the most common character in a string.
sample input :
var = 'Hello world '
sample output :
result : l, because l is repeating 3 times]





def most_common_character(self):
    char_counts = {}
    for char in self:
        char_counts[char] = char_counts.get(char,0)+1
    max_count = 0
    most_common_char = None
    for char,count in char_counts.items():
        if count>max_count:
            max_count = count
            most_common_char = char
    return f'most common character is {most_common_char} because it is repeating {max_count} times'
print(most_common_character('Hello world'))
````

````python



# Question 5: Write a Python program to check if two strings are anagrams.
sample input :
var1 = 'listen'
var2 = 'silent'
sample output :
result : both are anagram because characters in var2 is in var1






def anagrams(x,y):
    if sorted(x) == sorted(y):
        return 'Both are anagram'
    else:
        return 'Not anagram'
print(anagrams('listen','silent'))
````

````python




# Question 6: Write a Python program to remove all the vowels from a string.
sample input :
var = 'python'
sample output :
result : pythn , because vowels are [AEIOUaeiou]





def remove_vowels(self):
    vowels = 'aeiouAEIOU'
    New_word = ''
    for i in self:
        if i not in vowels:
            New_word += i
    return New_word
print(remove_vowels('python'))
````

````python


 Question 8: Write a Python program to capitalize the first letter of each word in a string [dont use built-in-function title].
sample input :
var = 'python is easy'
sample output :
result : Python Is Easy





def title_fun(self):
    words = self.split()
    New_list = []
    for i in words:
        if i :
            fun = i[0].upper()+i[1:].lower()
            New_list.append(fun)
            New_fun = ' '.join(New_list)
    return New_fun
print(title_fun('python is easy'))

````
