```
# Question 1: Write a function that takes two sets as input and returns their union.
Sample Input: set1 = {1, 2, 3}, set2 = {3, 4, 5}
Sample Output: {1, 2, 3, 4, 5}
```



````python
# ans:
set1 = {1,2,3}
set2 = {3,4,5}
set3 = set1.union(set2)
print(set3)
````

```
# Question 7 : Write a function that returns the number of elements in a set.
Sample Input: my_set = {10, 20, 30, 40, 50}
Sample Output: 5
```



````python
# ans:
def No_sets(self):
  new_set = 0
  for i in self:
    new_set +=1
  return new_set
print(No_sets({10,20,30,40,50}))
````

```
# Question 10:Write a function that generates the powerset of a given set (all possible subsets).
Sample Input: my_set = {1, 2}
Sample Output: {{}, {1}, {2}, {1, 2}}
```



````python
# ans:
def generating_powerset(self):
  powerset = []
  for i in range(2**len(self)):
    subset = []
    for j in range(len(self)):
      if (i and(1<<j)) !=0:
        subset.append(self[j])
    powerset.append(subset)
  return powerset
print(generating_powerset({1,2}))
````

````python
def powerset(self):
  if not self:
    return [[]]
  x = self.pop()
  subsets = powerset(self)
  return subsets + [subset + [x] for subset in subsets]
print(powerset({1,2}))
````
